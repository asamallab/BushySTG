# Generation of Boxplots individually

```r
## Here Boxplots of G-density for 4 models are being generated (those have been shown in the main text).
## To get the boxplots of SI use the data for associated network structure.
## Same code can be used to generate the boxplots for 'average indegree of non-GoE states' and 'average convergence rate'
## by taking the corresponding arguments for 'y' (Note: For average convergence rate take the inverse of 
## average transient lengths present in the dataframes)

library(tidyverse)
library(colorspace)
library(rcartocolor)
library(ggdist)
library(ggridges)
library(ggbeeswarm)
library(gghalves)
library(patchwork)
library(latex2exp)
library(patchwork)
library(ggplot2)

bio_data_df <-read.delim("../data/all_dataframes/bio_basinprops_main_models.tsv", sep='\t')

## ORIGIGNAL VERSION

theme_set(theme_void())

theme_update(
  axis.text.x = element_text(color = "black", size = 12, 
                             margin = margin(t = 6)),
  axis.text.y = element_text(color = "black", size = 12, hjust = 1, 
                             margin = margin(r = 6), angle = 45, vjust = 0.5),
  axis.line.x = element_line(color = "black", linewidth = 0.5),
  axis.line.y = element_line(color = "black", linewidth = 0.5),
  axis.ticks.x = element_line(linewidth = 0.5), 
  axis.ticks.y = element_line(linewidth = 0.5), 
  axis.ticks.length.y = unit(.125, "cm"),
  axis.ticks.length.x = unit(.125, "cm"),
  axis.title.y = element_text (angle = 90, vjust = 0.5, size = 14, margin = margin(t = 7)),
  axis.title.x = element_text (angle = 0, vjust = 0.5, size = 14, margin = margin(t = 7)),
  panel.grid.major.x = element_line(color = "grey90", linewidth = .6),
  panel.grid.major.y = element_line(color = "grey90", linewidth = .6),
  plot.background = element_rect(fill = "white", color = "white"),
  plot.margin = margin(0, 0, 1.3, 0.7, "cm"),
  plot.title = element_text(size = 16, face='bold', margin = margin(t = 7), hjust = 0.5)
)

## custom colors
my_pal <- rcartocolor::carto_pal(n = 8, name = "Bold")[c(1, 3, 7, 2)]

model_name <- "buylla"
dataset <- sprintf("../data/all_dataframes/%s_basin_props_plot_df.tsv", model_name)
data <-read.delim(dataset, sep='\t')
data$func_type <- factor(data$func_type, levels=c("sc-NCF", "sc-RoF", "sc-EUF", "EF"))


buylla <- ggplot(data, aes(x = func_type, y = g_density , color = func_type, fill = func_type)) + 
  geom_hline(yintercept = bio_data_df[bio_data_df$models ==model_name,]$g_density, linetype = 'dashed', linewidth = 0.5) + 
  scale_color_manual(values = my_pal, guide = "none") +
  scale_fill_manual(values = my_pal, guide = "none") +
  stat_boxplot(geom = "errorbar", width = 0.2, linewidth = 0.5) +
  geom_boxplot(lwd=0.5, alpha = .5, size = 1.5, width= 0.5, position = position_dodge(width=0.9), outlier.size=0.5) + 
  stat_summary(fun = "mean", geom="point", size=1.5, color="black") + 
  xlab(expression(bold("Type of BF"))) +
  ylab(TeX("\\textbf{\\textit{G}-density}")) +
  scale_x_discrete(label=c("NCF", "RoF", "EUF", "EF")) + 
  scale_y_continuous(limits = c(0.6, 1))+ 
  ggtitle("RSCN-GRN") + 
  coord_flip()

#buylla

ggsave("../data/boxplots/G_density/Main_figures/01_buylla_Rplot.png", width = 6, height = 3.9, dpi=600)

model_name <- "sullivan"
dataset <- sprintf("../data/all_dataframes/%s_basin_props_plot_df.tsv", model_name)
data <-read.delim(dataset, sep='\t')
data$func_type <- factor(data$func_type, levels=c("sc-NCF", "sc-RoF", "sc-EUF", "EF"))


sullivan <- ggplot(data, aes(x = func_type, y = g_density , color = func_type, fill = func_type)) + 
  geom_hline(yintercept = bio_data_df[bio_data_df$models ==model_name,]$g_density, linetype = 'dashed', linewidth = 0.5) + 
  scale_color_manual(values = my_pal, guide = "none") +
  scale_fill_manual(values = my_pal, guide = "none") +
  stat_boxplot(geom = "errorbar", width = 0.2, linewidth = 0.5) +
  geom_boxplot(lwd=0.5, alpha = .5, size = 1.5, width= 0.5, position = position_dodge(width=0.9), outlier.size=0.5) + 
  stat_summary(fun = "mean", geom="point", size=1.5, color="black") + 
  xlab(expression(bold("Type of BF"))) +
  ylab(TeX("\\textbf{\\textit{G}-density}")) +
  scale_x_discrete(label=c("NCF", "RoF", "EUF", "EF")) + 
  scale_y_continuous(limits = c(0.8, 1))+ 
  ggtitle("EMT-GRN") + 
  coord_flip()

#sullivan

ggsave("../data/boxplots/G_density/Main_figures/02_sullivan_Rplot.png", width = 6, height = 3.9, dpi=600)


model_name <- "cortical"
dataset <- sprintf("../data/all_dataframes/%s_basin_props_plot_df.tsv", model_name)
data <-read.delim(dataset, sep='\t')
data$func_type <- factor(data$func_type, levels=c("sc-NCF", "sc-RoF", "sc-EUF", "EF"))



cortical <- ggplot(data, aes(x = func_type, y = g_density , color = func_type, fill = func_type)) + 
  geom_hline(yintercept = bio_data_df[bio_data_df$models ==model_name,]$g_density, linetype = 'dashed', linewidth = 0.5) + 
  scale_color_manual(values = my_pal, guide = "none") +
  scale_fill_manual(values = my_pal, guide = "none") +
  stat_boxplot(geom = "errorbar", width = 0.2, linewidth = 0.5) +
  geom_boxplot(lwd=0.5, alpha = .5, size = 1.5, width= 0.5, position = position_dodge(width=0.9), outlier.size=0.5) + 
  stat_summary(fun = "mean", geom="point", size=1.5, color="black") + 
  xlab(expression(bold("Type of BF"))) +
  ylab(TeX("\\textbf{\\textit{G}-density}")) +
  scale_x_discrete(label=c("NCF", "RoF", "EUF", "EF")) + 
  scale_y_continuous(limits = c(0, 1))+ 
  ggtitle("NCD-GRN") + 
  coord_flip()

#cortical

ggsave("../data/boxplots/G_density/Main_figures/03_cortical_Rplot.png", width = 6, height = 3.9, dpi=600)

model_name <- "cardiac"
dataset <- sprintf("../data/all_dataframes/%s_basin_props_plot_df.tsv", model_name)
data <-read.delim(dataset, sep='\t')
data$func_type <- factor(data$func_type, levels=c("sc-NCF", "sc-RoF", "sc-EUF", "EF"))



cardiac <- ggplot(data, aes(x = func_type, y = g_density , color = func_type, fill = func_type)) + 
  geom_hline(yintercept = bio_data_df[bio_data_df$models ==model_name,]$g_density, linetype = 'dashed', linewidth = 0.5) + 
  scale_color_manual(values = my_pal, guide = "none") +
  scale_fill_manual(values = my_pal, guide = "none") +
  stat_boxplot(geom = "errorbar", width = 0.2, linewidth = 0.5) +
  geom_boxplot(lwd=0.5, alpha = .5, size = 1.5, width= 0.5, position = position_dodge(width=0.9), outlier.size=0.5) + 
  stat_summary(fun = "mean", geom="point", size=1.5, color="black") + 
  xlab(expression(bold("Type of BF"))) +
  ylab(TeX("\\textbf{\\textit{G}-density}")) +
  scale_x_discrete(label=c("NCF", "RoF", "EUF", "EF")) + 
  scale_y_continuous(limits = c(0.85, 1))+ 
  ggtitle("EHD-GRN") + 
  coord_flip()

#cardiac

ggsave("../data/boxplots/G_density/Main_figures/04_cardiac_Rplot.png",  width = 6, height = 3.9, dpi=600)
```

# Combining the boxplots

```r
library("png")
library("ggplot2")

files <- list.files(path="../data/boxplots/G_density/Main_figures/", pattern="*.png", all.files=T, full.names=T)
filelist <- lapply(files, readPNG)
names(filelist) <- paste0(basename((files)))
list2env(filelist, envir=.GlobalEnv)

par(mar=rep(0,4))
layout(matrix(1:length(names(filelist)), ncol=2, byrow=TRUE))

for(i in 1:length(names(filelist))){
  img <- readPNG(files[[i]])
  plot(NA,xlim=0:1,ylim=0:1,bty= "n",axes =0,xaxs="i",yaxs="i") + theme(aspect.ratio=3/2)
  rasterImage(img,0,0,1,1)
}
dev.off()
```
